# IMAGE TO PENCIL SKETCH USING PYTHON

IMPORT LIBRARIES

```python
import cv2
import matplotlib.pyplot as plt
import os
```

LOAD IMAGE USING OPENCV LIBRARY

```python
image=cv2.imread("image.jpg")
```

SET SIZE OF IMAGE 

```python
resized_img = cv2.resize(image, (1000, 700), interpolation=cv2.INTER_AREA)
```

```python
cv2.imwrite('resized_image.jpg', resized_img)
```

SHOW RESIZED IMAGE

```python
cv2.imshow('Image', resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

SHOWING IMAGE BY USING MATPLOTLIB 

```python
plt.imshow(resized_img)
plt.axis(False)
plt.show()
```

```python
plt.imshow(resized_img[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
RGB_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

CONVERT RGB IMAGE TO GREY IMAGE

```python
grey = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
```

INVERT GRAY IMAGE

```python
invert_img = cv2.bitwise_not(grey)
```

CONVERT GRAY IMAGE TO BLUR IMAGE

```python
blur = cv2.GaussianBlur(invert_img, (111,111),0)
```

CONVERT BLUR IMAGE TO INVERTED BLUR IMAGE

```python
invblur_img = cv2.bitwise_not(blur)
```

NOW WE CAN COMBINE GREY IMAGE AND INVERTED BLUR IMAGE FOR SKETCH 

```python
sketch = cv2.divide(grey ,invblur_img, scale=256.0)
```

```python
cv2.imwrite('sketch.png', sketch)
```

SHOWING SKETCH IMAGE

```python
cv2.imshow('sketch image',sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

COMAPRE ORIGINAL IMAGE TO SKETCH IMAGE

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
